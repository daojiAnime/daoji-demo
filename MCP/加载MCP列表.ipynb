```python
import rich
from fastmcp.client import Client
from fastmcp.client.transports import MCPConfigTransport

config: MCPConfigTransport = {
    "mcpServers": {
        "context7": {"url": "https://mcp.context7.com/mcp"},
        "deepwiki": {"url": "https://mcp.deepwiki.com/mcp"},
    }
}
client = Client(config)

async with client:
    tools = await client.list_tools()
    rich.print(tools)
```

```python
async with client:
    response = await client.call_tool("context7_resolve-library-id", {"libraryName": "fastmcp"})
    rich.print(response.content[0].text)
```

```python
mcp_config = {
    "mcpServers": {
        "mcp-server-hotnews": {"command": "npx", "args": ["-y", "@wopal/mcp-server-hotnews"]},
        "browser": {"url": "http://127.0.0.1:8089/sse"},
    }
}
client = Client(mcp_config)

async with client:
    tools = await client.list_tools()
    rich.print(tools)
    response = await client.call_tool("mcp-server-hotnews_get_hot_news", {"sources": [1]})
    rich.print(response.content[0].text)
```
